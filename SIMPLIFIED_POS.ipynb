```
def tagset_mapping(source_tag):

  tagset_mapping1= {'NN':'NN','NNP':'NN','NST':'NN','NNC':'NN','NNPC':'NN',
                 'PR':'PR','PRP':'PR','PRF':'PR','PRL':'PR','PRC':'PR','PRQ':'PR','PRI':'PR',
                 'DM':'DM','DMD':'DM','DMQ':'DM','DMI':'DM','DEM':'DM',
                 'VM':'VM','VAUX':'VM',
                 'JJ':'JJ',
                 'RB':'RB',
                 'PSP':'PSP',
                 'CC':'CC','CCD':'CC','CCS':'CC',
                 'RP':'RP','RPD':'RP','INJ':'RP','INTF':'RP','NEG':'RP',
                 'QT':'QT','QTF':'QT','QTC':'QT','QTO':'QT','QC':'QT',
                 'RD':'RD','RDF':'RD','SYM':'RD','PUNC':'RD','UNK':'RD','ECH':'RD','X':'RD'
                                          }
  for k,v in tagset_mapping1.items():
      if(source_tag==k):
        source_tag=tagset_mapping1[k]
  return source_tag
```

```
import nltk
nltk.download('punkt')
from nltk import word_tokenize
from nltk.translate.bleu_score import sentence_bleu
```

```
t=[]
translated_file = "/content/URMT_100_OP_pos.txt"
f5 = open('/content/URMT_100_OP_pos_simp.txt', "a+")
with open(translated_file, "r") as f1:
  lines1 = [line.rstrip('\n') for line in f1]
  #print(lines1)
 
  #print(lines2)
  for l1 in lines1:
    t=[]
    tokens_translate=word_tokenize(l1)
    for s in tokens_translate:
      t.append(tagset_mapping(s))
    tokens_translate=t
    for to in tokens_translate:
      f5.write(str(to)+" ")
    f5.write("\n")
f5.close()
```
